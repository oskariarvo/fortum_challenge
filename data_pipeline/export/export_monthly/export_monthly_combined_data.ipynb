```python
catalog = "fortum_challenge_data"
schema = "04_results"
table = "consumption_monthly_combined"

df = spark.table(f"{catalog}.{schema}.{table}")


ordered_df = (
    df
    .orderBy("measured_at")
    .coalesce(1)
)

ordered_df.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("/Volumes/fortum_challenge_data/04_results/exports/monthly/consumption_monthly_combined_csv")
```
