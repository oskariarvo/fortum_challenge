```python
catalog = "fortum_challenge_data"
schema = "04_results"
table = "consumption_hourly_actuals"

df = spark.table(f"{catalog}.{schema}.{table}")

ordered_df = (
    df
    .orderBy("group_id", "timestamp_utc")
    .coalesce(1)
)

ordered_df.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("/Volumes/fortum_challenge_data/04_results/exports/hourly/consumption_hourly_actuals_csv")
```
