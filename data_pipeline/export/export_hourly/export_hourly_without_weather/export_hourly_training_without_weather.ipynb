```python
catalog = "fortum_challenge_data"
schema = "03_gold"
table = "hourly_training_dataset_without_weather"

df = spark.table(f"{catalog}.{schema}.{table}")


ordered_df = (
    df
    .orderBy("group_id", "timestamp_utc")
    .coalesce(1)
)

ordered_df.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("/Volumes/fortum_challenge_data/03_gold/exports/hourly_without_weather/hourly_training_dataset_without_weather_csv")
```
